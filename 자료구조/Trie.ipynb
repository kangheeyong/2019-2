```python
def make_dic(words) :
    tree_front = {}
    tree_rear = {}
    
    for word in words :
        size = len(word)
        tree_front[size] = tree_front.get(size,{})
        curr = tree_front[size] 
        for ch in word :
            curr['count'] = curr.get('count',0) + 1
            curr[ch] = curr.get(ch,{}) 
            curr = curr[ch]
        
        tree_rear[size] = tree_rear.get(size,{})
        curr = tree_rear[size] 
        for ch in word[::-1] :
            curr['count'] = curr.get('count',0) + 1
            curr[ch] = curr.get(ch,{}) 
            curr = curr[ch]
            
    return [tree_front, tree_rear]

def find(tree, query) :
    
    size = len(query)
    tree_front,tree_rear = tree[0], tree[1]
    
    
    if query[0] != '?' :    
          
        if size not in tree_front :
            return 0
        curr = tree_front[size]           
        
        for ch in query :
            if ch == '?' :
                return  curr['count']
            
            if ch not in curr : 
                return 0                
            curr = curr[ch]
        
    else :
        if size not in tree_rear :
            return 0
        curr = tree_rear[size]
        for ch in query[::-1] :
            if ch == '?' :
                return curr['count']
                    
            if ch not in curr : 
                return 0            
            curr = curr[ch]
        
    
def solution(words, queries):
    
    tree = make_dic(words)
    
    answer = []
    for query in queries :
        answer += [find(tree,query)]
        
    #print(answer)
    return answer
    
```
