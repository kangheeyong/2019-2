```python
#1

def solve(s,gap) :
    L = len(s)
    data = []
    
    before = ''
    for idx in range(0,L,gap) : 
        if before != s[idx : idx + gap] :
            data += [[s[idx : idx + gap],1]]
        else :
            data[-1][1] +=1
        before = s[idx : idx + gap]
    
    answer = ''
    for ch, cnt in data : 
        if cnt == 1 :
            answer += ch
        else :
            answer += (str(cnt)+ch)
    #print(gap,answer)
    return len(answer)
def solution(s):
    L = len(s)
    _min = L
    
    for i in range(1,L//2+1) : # 이거 다시 체크
        _min = min(solve(s,i),_min)
    
    
    answer = _min
    return answer
```

```python
def change_stack(data,cnt) :
    temp =[]
    h,w = len(data), len(data[0])
    
    for y in range(0,h) :
        for x in range(0,w) :
            if data[y][x] == cnt :
                temp += [[y,x]]
    return temp

def move(data,dy,dx) :
    temp =[]
    for y,x in data :
        temp += [[y+dy,x+dx]]
    return temp

def lotate(data,L) :
    temp =[]
    for y,x in data :
        temp += [[L-1-x, y]]
    return temp

def check(keys,locks) :
    cnt = 0
    
    for key in keys :
        
        if key in locks :
            cnt +=1
    #print(cnt,keys,locks)
    if cnt == len(keys) and len(keys) == len(locks) :
        return True
    else :
        return False
    
def solution(key, lock):
    
    M, N = len(key), len(lock)
    key_, lock_ = change_stack(key,1), change_stack(lock,0)
    
    #print(key_,lock_)
    for __ in range(0,4) :
        key_ = lotate(key_,M)
        for dx in range(-M+1,N+M-1) :
            for dy in range(-M+1,N+M-1) :
                temp_keys = move(key_,dy,dx)
                temp_keys = list(filter(lambda x : 0<=x[0]<N and 0<=x[1]<N,temp_keys))
                #print(temp_keys,check(temp_keys,lock_),lock_)
                if check(temp_keys,lock_) :
                    return True
    
    
    

    return False
```

```python
#4

def make_trie(words) : # 요놈 구조 다시 보기
    
    trie = {}
    for word in words :
        size = len(word)
        trie[size] = trie.get(size,{})
        curr = trie[size] # 나중에 체크
        for ch in word :
            curr['count'] = curr.get('count',0)+1
            curr[ch] = curr.get(ch,{})
            curr = curr[ch] 
            
    return trie
    

def solution(words, queries):
    answer = []
    # 역방향 잘못만듬
    trie_front, trie_rear = make_trie(words),make_trie(words[::-1])
    print(trie_rear[5])
    for query in queries :
        if query[0] == '?' :
            curr = trie_rear
            data = query[::-1]
        else : 
            curr = trie_front
            data = query
        
        answer += [0]  
        if len(data) in curr :
            curr = curr[len(data)]
            for d in data :
                if d == '?' :
                    answer[-1] = curr['count']
                if d not in curr :
                    break
                curr = curr[d]
                
    
    return answer
```

```python
#5 

# 문제와 답을 외운 케이스 문제 분석 다시
def check(x,y,a,answer) :
    if a == 0 :
        cnd1 = y==0 or [x,y-1,0] in answer
        cnd2 = [x,y,1] in answer or [x-1,y,1] in answer
        if cnd1 or cnd2 :
            return True
        else :
            return False
    else :
        cnd1 = [x,y-1,0] in answer or [x+1,y-1,0] in answer
        cnd2 = [x-1,y,1] in answer and [x+1,y,1] in answer # 숫자실수 발생
        if cnd1 or cnd2 :
            return True
        else :
            return False

def solution(n, build_frame):
    # [x,y,a,b]
    # a==0 : 기둥, a == 1 : 보
    # b == 0 : 삭제, b == 1 : 설치
    answer = []
    
    for x,y,a,b in build_frame : 
        if b == 1 :
            if check(x,y,a,answer) : 
                answer.append([x,y,a])
        else :
            answer.remove([x,y,a])
            for _x,_y,_a in answer :
                if not check(_x,_y,_a,answer) :
                    answer.append([x,y,a])
                    break
    answer.sort()
    #print(answer)
    return answer
```
