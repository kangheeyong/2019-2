## 임의 데이터 생성

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# https://namyoungkim.github.io/probability/statistics/2017/09/11/probability_2/
mean_1 = np.array([-1,-1])
variance_1 = np.array([[0.1,0],[0,0.1]])

data_1 =  np.random.multivariate_normal(mean_1, variance_1, 1000)

mean_2 = np.array([1,1])
variance_2 = np.array([[1.4,-0.6],[-0.6,0.4]])

data_2 =  np.random.multivariate_normal(mean_2, variance_2, 1000)



data = np.concatenate((data_1,data_2),axis = 0)
```

```python
plt.scatter(data[:,0],data[:,1],c=np.array([[0,0,0]]),s =1)
```
